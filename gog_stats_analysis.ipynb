```python
import pandas as pd
import chart_studio.plotly as py
import plotly.express as px
import plotly.graph_objects as go
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Prepare Data
%run src/gog_stats_exporter.py -d data/galaxy-2.0.db -o data/output.csv -p
```

```python
df = pd.read_csv("data/output.csv")
df
```

```python
stats = pysqldf('''SELECT
                       platform as Platform, count(`title.title`) as `Number of Games`, 
                       round(sum(`minutesInGame`)/60.0, 2) as `Hours Total`, 
                       sum(`myAchievementsCount.unlocked`) as `Achievements Total`,
                       round(sum(`price.current`), 2) as `Value Current`,
                       round(sum(`price.high`), 2) as `Value New`
                    FROM df GROUP BY platform ORDER BY `Hours Total` DESC''')

fig = go.Figure(data=[go.Table(
    header=dict(values=list(stats.columns), align='left'),
    cells=dict(values=[stats['Platform'],stats['Number of Games'],stats['Hours Total'],stats['Achievements Total'],stats['Value Current'],stats['Value New']], align='left'))
])
fig.show()
```

```python
df2 = pysqldf('''SELECT
                     `title.title` as Title,
                     `minutesInGame`/60.0 as Hours
                 FROM df WHERE Hours > 0 ORDER BY Hours''')

fig = px.bar(df2, x='Hours', y='Title', orientation='h', text='Hours', title='Hours playtime')
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    height=20*df2['Title'].count(),
)
fig.show()
```

```python
df3 = pysqldf('''SELECT
                     `title.title` as Title, 
                     100.0 * `myAchievementsCount.unlocked` / `myAchievementsCount.all` as Achievements 
                 FROM df WHERE Achievements > 0 ORDER BY Achievements''')

fig = px.bar(df3, x='Achievements', y='Title', orientation='h', text='Achievements', title='Achievemets percentage')
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(
    height=20*df3['Title'].count(),
)
fig.show()
```
